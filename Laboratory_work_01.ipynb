<br/>
<b>Explore the following code snippets to understand how does reshape() and resize() work, and move to the tasks</b>

```python
A = np.arange(10)
print("Original array:", A)
A = A.reshape(5,2)
print("\nReshaped array:")
print(A)
```

```python
B = np.arange(10)
print("Original array:", B)
B.resize(5,2)
print("\nReshaped array:")
print(B)
```

<span style="font-size:15px;"><b>TASK 5:</b></span>  Array unions 

<span style="font-size:20px; color:blue">S</span>ometimes you need to create an array as result of the union. The following code demonstrates simple union or staking of arrays. Note, that your array should be the same dimensions to be able to stak them

```python
A = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
B = A[::-1]
C = A[:, ::-1]
print(A)
print("\n",B)
print("\n",C)
D = np.stack((A, B, C))
D
```
