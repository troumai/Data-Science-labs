```python
from random import randint
```

```python
def simple_random_sampling(data, size):
    population=[]
    for i in range(size):
         population.append((data.iloc[randint(0,len(data.index)-1)]))
    return pd.DataFrame(population, columns=data.columns)
```

```python
simple_random_sampling(data, 6)
```

    <br/> Remember, you are not allowed to use any functions from scipy, or sklearn or any other packages.

    </span>

```python
def systematic_sampling(data, step):
    indexes = np.arange(0,len(data),step=step)
    systematic_sample = data.iloc[indexes]
    return systematic_sample
```

```python
systematic_sampling(data, 6)
```

 <h3>TASK 1.4 </h3> <br/>Stratified sampling -  write a function (don't use a any module from scikit-learn or stats packages) that creates a stratified sample from you <code>dataset</code>
    <br/> Remember, you are not allowed to use any functions from scipy, or sklearn or any other packages.
    <br/> <span style="color:#00a0e5">(  -  ,        .          ) </span>
    </span>

<p style="text-align:right; color:#1460AA;">+10
    points </p>

```python
def stratified_sampling(data, column, sample_size):
    new_sample = pd.DataFrame()
    for i in data[column].unique():
        percent = len(data[data[column]==i])/len(data)*100
        sample = data[data[column] == i].sample(int(percent/100*sample_size))
        new_sample = pd.concat([new_sample, sample], ignore_index=True)
    return new_sample
```

```python
stratified = stratified_sampling(data,'hotel', 10000)
stratified
```

```python
# here mu-population mean, sigma-population std, n-sample size, m-samples amount
def SimulateSamplee(mu, sigma, n, m):
    means = []
    for j in range(m):
        xs = np.random.normal(mu, sigma, n)
        means.append(np.mean(xs))
    return means
```

```python
adults_mu = data.adults.mean()
adults_sigma = data.adults.std()
```

```python
sample_means_adults = SimulateSamplee(adults_mu, adults_sigma, 100, 100)
```

Sampling can be particularly useful with data sets that are too large to efficiently analyze in full. Identifying and analyzing a representative sample is more efficient and cost-effective than surveying the entirety of the data or population.
An important consideration, though, is the size of the required data sample and the possibility of introducing a sampling error. In some cases, a small sample can reveal the most important information about a data set. In others, using a larger sample can increase the likelihood of accurately representing the data as a whole, even though the increased size of the sample may impede ease of manipulation and interpretation.

Simple random sampling: Software is used to randomly select subjects from the whole population.

Systematic sampling: A sample is created by setting an interval at which to extract data from the larger population -- for example, selecting every 10th row in a spreadsheet of 200 items to create a sample size of 20 rows to analyze.

Stratified sampling: Subsets of the data sets or population are created based on a common factor, and samples are randomly collected from each subgroup.

Estimation is the process of determining the numerical value of uncertain population values based on missing data, such as a sample. We used estimation in task 2 and compared it to any functions.

The Central Limit Theorem states that as the sample size grows larger, the sampling distribution of the sample implies approaches a normal distribution, regardless of the form of the population distribution.

The shape of the sampling distribution depends on the sample size. If you perform the study using the same procedure and change only the sample size, the shape of the sampling distribution will differ for each sample size. As sample size increases, the sampling distribution more closely approximates the normal distribution, and the spread of that distribution tightens.
